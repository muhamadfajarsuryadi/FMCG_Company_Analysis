```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load data final
df_fmcg = pd.read_csv('master_data_final.csv')
df_fmcg.columns = df_fmcg.columns.str.strip()

# 2. Plotting Price Elasticity Regression
plt.figure(figsize=(12, 5))
sns.regplot(
    data=df_fmcg, x='unit_price', y='quantity',
    scatter_kws={'alpha': 0.25, 'color': '#1f77b4'},
    line_kws={'color': '#d62728', 'linewidth': 2}
)
plt.title('Statistical Analysis: Hubungan Price vs Quantity', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd

# 1. Load data final
df_fmcg = pd.read_csv('master_data_final.csv')
df_fmcg.columns = df_fmcg.columns.str.strip()

# 2. Cross-Category Product Ranking for Bundling
print("Diversitas transaksi tinggi. Berikut Top Produk per Kategori untuk Bundling:")
product_ranking = df_fmcg.groupby(['category', 'product_name'])['quantity'].sum().reset_index()

top_bundle_candidates = (
    product_ranking.sort_values(['category', 'quantity'], ascending=[True, False])
    .groupby('category').head(2)
)
print(top_bundle_candidates.to_string(index=False))
```
